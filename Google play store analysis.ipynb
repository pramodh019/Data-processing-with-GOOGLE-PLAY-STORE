```python
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv(r"C:\Users\pramo\Desktop\jupyter project\Data processing with google play store\googleplaystore.csv")
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
impute.fit(data.iloc[:, :])  # Apply iloc to DataFrame
data.iloc[:, :] = impute.transform(data.iloc[:, :])  # Apply iloc to DataFrame
```
